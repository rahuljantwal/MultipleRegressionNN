### Implementing Simple Linear Regression using a single Neuron

```python
X = np.array(bike_df[['atemp','hum','windspeed']])
Y = np.array(bike_df['cnt'])
```

#### Adding Bias Term

```python
bias = np.ones(shape=(len(X),1))
X = np.append(bias, X, axis=1)
```

### Cost Function

```python
def cost_function(X, Y, weight):
    total_error = 0.0
    total_error = np.mean((Y - np.dot(X,weight))**2)
    return total_error
```

### Updating Weights or Coefficients

```python
def update_weights(X, Y, weight, learning_rate):
    err = (Y - np.dot(X,weight))
    weight -= (-2*np.dot(X.T, err)/ len(X))*learning_rate
    return weight
```

### Training Model

```python
def train(X, Y, weight, learning_rate, iters):
    cost_history = []
    weight_history = []
    bias_history = []
    for i in range(iters):
        weight = update_weights(X, Y, weight, learning_rate)
        #Calculate cost for auditing purposes
        cost = cost_function(X, Y, weight)
        cost_history.append(cost)
        weight_history.append(weight)
        #bias_history.append(bias)
        # Log Progress
        if i % 1000 == 0:
            print ("iter: "+str(i) + " cost: "+str(cost) + " weight: "+str(weight))

    return weight, cost_history, weight_history
```

#### Training log

```python
w = np.array([0., 0., 0., 0.])
up_weight, cost_hist, weight_hist = train(X,Y,w,0.05,20001)
```

### Compare results with Tensorflow using Keras

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(1,kernel_initializer="zeros",input_shape=(3,)))
model.compile(loss='mean_squared_error',
                  optimizer="sgd",
                  metrics=['mae'])
```

```python
X = np.array(bike_df[['atemp','hum','windspeed']])
Y = np.array(bike_df['cnt'])
```

```python
hist = model.fit(X,Y, epochs=2000, verbose=0)
```
