### Implementing Simple Linear Regression using a single Neuron

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
import imageio
```

```python
os.chdir("/Users/rahuljantwal/Desktop/git/MultipleRegressionNN")
bike_df = pd.read_csv("./data/Bike-Sharing-Dataset/day.csv",delimiter=',')
N = np.shape(bike_df)[0]
```

```python
X = bike_df[['atemp','hum','windspeed']]
Y = bike_df['cnt']
```

### Cost Function

```python
def cost_function(X, Y, weight, bias):
    total_error = 0.0
    for i in range(len(X)):
        total_error += (Y[i] - (weight*X[i] + bias))**2
    return total_error / len(X)
```

### Implementing Gradient Descent

```python
def update_weights(X, Y, weight, bias, learning_rate):
    weight_deriv = 0
    bias_deriv = 0

    for i in range(len(X)):
        # Calculate partial derivatives
        # -2x(y - (mx + b))
        weight_deriv += -2*X[i] * (Y[i] - (weight*X[i] + bias))

        # -2(y - (mx + b))
        bias_deriv += -2*(Y[i] - (weight*X[i] + bias))

    # We subtract because the derivatives point in direction of steepest ascent
    weight -= (weight_deriv / len(X)) * learning_rate
    bias -= (bias_deriv / len(X)) * learning_rate

    return weight, bias
```

### Training Model

```python
def train(X, Y, weight, bias, learning_rate, iters):
    cost_history = []
    weight_history = []
    bias_history = []
    for i in range(iters):
        weight,bias = update_weights(X, Y, weight, bias, learning_rate)
        #Calculate cost for auditing purposes
        cost = cost_function(X, Y, weight,bias)
        cost_history.append(cost)
        weight_history.append(weight)
        bias_history.append(bias)
        # Log Progress
        if i % 500 == 0:
            print ("iter: "+str(i) + " cost: "+str(cost) + " weight: "+str(weight) + " bias: "+str(bias))

    return weight, bias, cost_history, weight_history, bias_history
```

#### Training log

```python
w = np.array([0, 0])
up_weight, up_bias, cost_hist, weight_hist, bias_hist = train(X,Y,w[0],w[1],0.05,6001)
```

```python
tt = np.linspace(np.min(bike_df['atemp']), np.max(bike_df['atemp']), 100)
bf_line = up_bias+up_weight*tt
```

#### Plot Linear Model

```python
plt.figure(figsize=(20,10))
plt.plot(bike_df['atemp'],bike_df['cnt'], 'ro', tt, bf_line, 'r-')
plt.xlabel('Temperature (Normalized)', fontsize=25)
plt.ylabel('Total Riders', fontsize=25)
plt.title('ANN Regression on 1D Bike Share data', fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.savefig('bike.png')
 
plt.show()
```

### Visualizing training with each iteration (a.k.a. Epoch)

```python
chart_pngs = []
for i in range(0,2000,5):
    plt.figure(figsize=(20,10))
    tt = np.linspace(np.min(bike_df['atemp']), np.max(bike_df['atemp']), 100)
    bf_line = bias_hist[i]+weight_hist[i]*tt
    plt.plot(bike_df['atemp'],bike_df['cnt'], 'ro', tt, bf_line, 'r-')
    plt.xlabel('Temperature (Normalized)', fontsize=18)
    plt.ylabel('Total Riders', fontsize=18)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.title('Regression Bike Share Riders', fontsize=18)
    plt.savefig('./charts/' + 'bike'+str(i)+'.png')
    #print(plt.get_fignums())
    plt.close("all")
    #plt.ioff()
#plt.show()
```

```python
chart_pngs = [f for f in os.listdir('./charts') if f.endswith('.png')]
import re
numbers = re.compile(r'(\d+)')

def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts

chart_pngs = sorted(chart_pngs,key=numericalSort)
```

```python
# Append all the charts (except the title slides)
pth = os.path.join(str(os.getcwd())+'/charts')
charts = []
for f in chart_pngs:
    charts.append(imageio.imread(pth+'/'+f))
```

```python
writer = imageio.get_writer('bikes.mp4', fps=60)
for chart in charts:
    writer.append_data(chart)
writer.close()
```
